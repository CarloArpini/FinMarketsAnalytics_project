```python
import yfinance as yf
from datetime import datetime
```

```python
start_date = datetime(2017, 6, 1)
end_date = datetime(2024, 7, 1)
```

```python
tickerslist=list(tickerset)
tickerslist.sort()
for i in range(len(tickerslist)):
    tickerslist[i]=tickerslist[i].replace('.', '-')
tickerslist
```

```python
data = yf.download(tickerslist[0], start=start_date, end=end_date)
data.drop(['Open', 'High', 'Close', 'Low', 'Volume'], axis=1, inplace=True)
data.reset_index(inplace=True)
data.rename({'Adj Close': tickerslist[0]}, axis='columns', inplace=True)
prices_stocks_list=pd.DataFrame(data)
# Loop through each ticker symbol
tickerslist.remove(tickerslist[0])
for ticker in tickerslist:
    # Download historical data from Yahoo Finance
    data = yf.download(ticker, start=start_date, end=end_date)
    data.drop(['Open', 'High', 'Close', 'Low', 'Volume'], axis=1, inplace=True)
    data.reset_index(inplace=True)
    data.drop('Date', inplace=True, axis=1)
    data.rename({'Adj Close': ticker}, axis='columns', inplace=True)
    # Append the data to the historical_data DataFrame
    prices_stocks_list=pd.concat([prices_stocks_list, data], axis=1)
# Reset index to have a clean DataFrame index
prices_stocks_list
```
